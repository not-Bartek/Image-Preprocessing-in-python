Resizing an image

```python
from PIL import Image
```

```python
img = Image.open('dog2.jpg')
img_resized = img.resize((200,200))
img_resized.save('dog_image_resized2.jpg')
```

```python
#displaying the image
img_res = mpimg.imread('dog_image_resized2.jpg')
img_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

**Converting RGB images to grayscale image using openCV**

```python
import cv2

img = cv2.imread('dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
grayscale_image.shape
```

```python
grayscale_image
```
